
# RayTune Experiments

```python
from ray import tune
```

```python
def objective(params):
    func = params['func']
    x = params['x']
    
    if func == 'second':
        return 2*x**2 - 6*x
    elif func == 'first':
        return 4*x**2+6*x-4
```

```python
def training_function(config):
    # Hyperparameters
    for step in range(10):
        # Iterative training function - can be any arbitrary training procedure.
        intermediate_score = objective(config)
        # Feed the score back back to Tune.
        tune.track.log(mean_accuracy=intermediate_score, accuracy = intermediate_score)
```

```python
analysis = tune.run(
    objective,
    config={
        "x": tune.uniform(-2, 5),
        "func": tune.grid_search(['first', 'second']),},
    num_samples = 100)
```

```python
analysis.get_all_configs()
```

```python
analysis.get_best_config(metric='_metric', mode='min')
```

```python
analysis.trial_dataframes['/Users/shaominliu/ray_results/objective_2021-05-24_15-54-12/objective_3a5df_00000_0_func=first,x=-0.36493_2021-05-24_15-54-12']
```

```python
def objective(step, alpha, beta):
    return (0.1 + alpha * step / 100)**(-1) + beta * 0.1


def training_function(config):
    # Hyperparameters
    alpha, beta = config["alpha"], config["beta"]
    for step in range(10):
        # Iterative training function - can be any arbitrary training procedure.
        intermediate_score = objective(step, alpha, beta)
        # Feed the score back back to Tune.
        tune.report(mean_loss=intermediate_score)


analysis = tune.run(
    training_function,
    config={
        "alpha": tune.grid_search([0.001, 0.01, 0.1]),
        "beta": tune.choice([1, 2, 3])
    })

print("Best config: ", analysis.get_best_config(
    metric="mean_loss", mode="min"))

# Get a dataframe for analyzing trial results.
df = analysis.results_df
```
